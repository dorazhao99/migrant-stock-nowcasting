# Facebook Data
Use the two functions below to query Facebook reach estimates. Requires an account token and Facebook Ad Account. Both are free.

```python
from facebook_business.api import FacebookAdsApi
from facebook_business.adobjects.adaccount import AdAccount
import json
import time
```

```python
# account token and account number
AT = '' # account token
ACT_NUMBER = '' # ad account id
```

```python
# account token and account number
at = AT
FacebookAdsApi.init(access_token=at, api_version='v9.0', debug=True)
my_account = AdAccount(ACT_NUMBER)
```

```python
# country codes
file = open('fbCountries.json')
countries = json.load(file)
countries = [x['country_code'] for x in countries]
```

```python
# for age-sex group estimates
def getFBReachAge(gender, age, expat):
    reach_estimate = []
    for country in countries:
        if expat: 
            targeting_spec = {
                'geo_locations':{
                    'countries':[country],
                  },
                'genders': [gender],
                'age_min': age['min'],
                'age_max': age['max'], 
                'behaviors':[{'id':6015559470583,'name':'Expats'}],

            }
        else:
            targeting_spec = {
                'geo_locations':{
                  'countries':[country],
                },
                'genders': [gender],
                'age_min': age['min'],
                'age_max': age['max'], 
            }
        params = {
          'targeting_spec': targeting_spec,
        }
        account_reach_estimate = my_account.get_reach_estimate(params=params)
        reach_estimate.append(account_reach_estimate[0]["users"])
```

```python
# for estimates not broken down by age-sex
def getFBReach(gender, expat):
    reach_estimate = []
    for country in countries:
        if expat: 
            targeting_spec = {
                'geo_locations':{
                    'countries':[country],
                  },
                'genders': [gender],
                'behaviors':[{'id':6015559470583,'name':'Expats'}],

            }
        else:
            targeting_spec = {
                'geo_locations':{
                  'countries':[country],
                },
                'genders': [gender],
            }
        params = {
          'targeting_spec': targeting_spec,
        }
        account_reach_estimate = my_account.get_reach_estimate(params=params)
        reach_estimate.append(account_reach_estimate[0]["users"])
    return reach_estimate
```

```python
# example query for the total expat estimates in a country
reach = getFBReach(0, True)
```
